```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
df = pd.read_csv("table-1.csv")
df = df.drop(df.filter(like='Unnamed', axis=1), axis=1)
df = df.drop("Row Number", axis=1)
df = df.drop(["16-44 years","45-59 years","60 + years"], axis=1)
df["Industries"] = pd.read_csv("table-1.csv")["Unnamed: 1"]
df.insert(0, 'Industries', df.pop('Industries'))
df = df.rename(columns={'With a disability': 'Median age With a disability', 'No disability': 'Median age No disability',
                       'With a disability.1': 'Percent working full-time, year-round With a disability', 'No disability.1': 'Percent working full-time, year-round No disability',
                       'With a disability.2': 'Median earnings of full-time, year-round workers With a disability', 'No disability.2': 'Median earnings of full-time, year-round workers No disability'})
for i, row in df.iterrows():
    try:
        df.loc[i, 'Median earnings of full-time, year-round workers With a disability'] = int(str(df.loc[i, 'Median earnings of full-time, year-round workers With a disability']).replace(',', ''))
        df.loc[i, 'Median earnings of full-time, year-round workers No disability'] = int(str(df.loc[i, 'Median earnings of full-time, year-round workers No disability']).replace(',', ''))
        df.loc[i, 'Percent working full-time, year-round With a disability'] = float(df.loc[i, 'Percent working full-time, year-round With a disability'])
        df.loc[i, 'Percent working full-time, year-round No disability'] = float(df.loc[i, 'Percent working full-time, year-round No disability'])
    except ValueError:
        pass
df = df.replace('-', np.nan)
df['Percent of workers in occupation with a disability'] = df['Percent of workers in occupation with a disability'].astype(float)
df['Median age With a disability'] = df['Median age With a disability'].astype(float)
df['Median age No disability'] = df['Median age No disability'].astype(float)
df['Median earnings of full-time, year-round workers With a disability'] = df['Median earnings of full-time, year-round workers With a disability'].astype(float)
df['Median earnings of full-time, year-round workers No disability'] = df['Median earnings of full-time, year-round workers No disability'].astype(float)
df['Percent working full-time, year-round With a disability'] = df['Percent working full-time, year-round With a disability'].astype(float)
df['Percent working full-time, year-round No disability'] = df['Percent working full-time, year-round No disability'].astype(float)
df = df.dropna()
df
```
